```python
from flask import Flask, render_template, request
from flask_socketio import SocketIO, emit

app = Flask(__name__)
socketio = SocketIO(app, cors_allowed_origins="*")

users = {}

@app.route('/')
def index():
    return render_template('index.html')

@socketio.on('new_user')
def handle_new_user(data):
    users[request.sid] = data['name']
    emit('receive_message', {
        'name': 'System',
        'message': f"{data['name']} joined the chat.",
        'side': 'left'
    }, broadcast=True)

@socketio.on('send_message')
def handle_message(data):
    name = users.get(request.sid, 'Unknown')
    emit('receive_message', {
        'name': name,
        'message': data['message'],
        'side': 'left'
    }, broadcast=True)

@socketio.on('offer')
def handle_offer(data):
    emit('offer', data, room=data['to'])

@socketio.on('answer')
def handle_answer(data):
    emit('answer', data, room=data['to'])

@socketio.on('ice_candidate')
def handle_ice_candidate(data):
    emit('ice_candidate', data, room=data['to'])

@socketio.on('disconnect')
def handle_disconnect():
    name = users.pop(request.sid, 'A user')
    emit('receive_message', {
        'name': 'System',
        'message': f"{name} left the chat.",
        'side': 'left'
    }, broadcast=True)

if __name__ == '__main__':
    socketio.run(app, host='0.0.0.0', port=8000)
```
